```python
def is_valid_ip(ip):
    p=ip.split(".")
    return check_part(p,0)
def check_part(p,i):
    if i==4:
        return len(p)==4
    if i>=len(p):
        return False
    part=p[i]
    if not(part.isdigit()and 0<=int(part)<=255):
        return False
    return check_part(p,i+1)
is_valid_ip("255.0.90.22")
```
